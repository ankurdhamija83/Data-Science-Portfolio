# 3. Data Cleaning

### Key date cleaning steps perfomed
- Drop columns with more than 98% ratio of a single value
- Replace `?` with NaN values
- Drop columns with high percentage of NaN values
- Replace NaN values with Unknown in select columns
- Drop tel_11, tel_10 and tel_9 columns
- Group low frequency values in categorical columns
- Remove rows with very few items - For example, Unknown/Invalid in Gender has only 2 values

**3.1 Get list of categorical columns**

```python
# Select categorical columns
catCols = df.select_dtypes("object").columns
```

```python
# View categorical columns
catCols
```

**3.2 Drop columns with high concentration of a single value**

```python
#Select columns with more 98% of similar values. These columns have low predictive power
filteredCols = filter(lambda col: (df[col].value_counts().sort_values(ascending=False)/df.shape[0])[0] >= 0.98, catCols)
```

```python
#We drop the following columns
colsToBeRemoved = list(filteredCols)
```

```python
print("Columns to be removed: ", colsToBeRemoved)
```

```python
#Delete colsToBeRemoved from the df
df = drop_col(df, colsToBeRemoved)
```

**3.3 Get an overview of the dataset again**

```python
#Overview of the dataset
#Get an overview of the Dataset
df_overview(df, max_cat=20, figsize=(10,5))
```

**3.4 Replace '?' with nan values**

```python
# replace ? with nan
df = df.replace('?',np.nan)
```

**3.5 Check for NaN values in the dataset**

```python
df.isnull().sum().sort_values(ascending=False)
```

**3.6 Drop weight column**

```python
#Delete colsToBeRemoved from the df
df = drop_col(df, ["weight"])
```

**3.9 Club values in a column**

```python
df.head()
```

```python
df.info()
```

```python
def groupVals(col_list, df):
    for col in col_list:
        valFreq = (df[col].value_counts())/df.shape[0]
#         print("THis is valFreq: ", valFreq)
        lessFreqVals = valFreq[valFreq<=.01]
#         print("\n\n")
#         print("These are values with lessFreq: ", lessFreqVals)        
        df.loc[df[col].isin(lessFreqVals.index.tolist()),col] = "Others"
        return df
```

```python
df = groupVals(["tel_1","tel_2"], df)
```

```python
df["tel_1"].value_counts()
```

**Remove rows with very few items - For example, Unknown/Invalid in Gender has only 2 values**

```python
df = df[~(df["gender"] == "Unknown/Invalid")]
```

# Model Building

### Logisitic Regression

```python
# Build Logistic Regression model
logisticModel = linear_model.LogisticRegression(penalty='l2', C=1)
logisticModel.fit(X_train, y_train)
```

```python
# Model Accuracy on the test set
print ("Logistic model Accuracy rate = %2.2f" % metrics.accuracy_score(y_test, logisticModel.predict(X_test)))
```

```python
# Model AUC on the test set
print ("Logistic Model AUC = %2.2f" % metrics.roc_auc_score(y_test, logisticModel.predict(X_test)))
```

```python
#Confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, logisticModel.predict(X_test))

# Visualize Confusion matrix
plt.figure(figsize=(9,9))
sns.heatmap(confusion_matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(metrics.accuracy_score(y_test, logisticModel.predict(X_test)))
plt.title(all_sample_title, size = 15);
```

```python
# Calculating Precision - From all the predicted 0 labels, how many did the model predict correctly
precision = confusion_matrix[0][0]/(confusion_matrix[0][0] + confusion_matrix[1][0])
print("LR Model Precision = %2.2f" % precision)
```

```python
# Calculating Sensitivity aka Recall - From all the actual 0 labels, how many did the model predict correctly
sensitivity = confusion_matrix[0][0]/(confusion_matrix[0][0] + confusion_matrix[0][1])
print("LR Model Sensitivity = %2.2f" % sensitivity)
```

```python
# Calculating Specificity - From all the actual 1 labels, how many did the model predict correctly
specificity = confusion_matrix[1][1]/(confusion_matrix[1][0] + confusion_matrix[1][1])
print("LR Model Specificity = %2.2f" % specificity)
```

We understand that the specificity of the model is very low. Out of all the spam emails, its predicting only 21% correctly.

```python
# Calculating Accuracy - Percentage of correct predictions by the model
accuracy = (confusion_matrix[0][0] + confusion_matrix[1][1])/(confusion_matrix[0][0] + confusion_matrix[0][1] + confusion_matrix[1][0] + confusion_matrix[1][1])
print("LR Model Accuracy = %2.2f" % accuracy)
```

```python
print("Classification report")
print(metrics.classification_report(y_test, logisticModel.predict(X_test)))
```
