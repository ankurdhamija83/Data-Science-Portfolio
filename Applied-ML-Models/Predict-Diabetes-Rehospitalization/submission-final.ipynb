# 3. Data Cleaning

### Key data cleaning steps perfomed
- Drop columns with more than 98% ratio of a single value
- Replace `?` with NaN values
- Drop columns with high percentage of NaN values
- Replace NaN values with Unknown in select columns
- Drop tel_11, tel_10 and tel_9 columns
- Group low frequency values in categorical columns
- Remove rows with very few items - For example, Unknown/Invalid in Gender has only 2 values

```python
def dataCleaning(df):
    
    df.set_index('encounter_id', inplace=True)
    
    # Select categorical columns
    catCols = df.select_dtypes("object").columns
    
    #Select columns with more 98% of similar values. These columns have low predictive power
    filteredCols = filter(lambda col: (df[col].value_counts().sort_values(ascending=False)/df.shape[0])[0] >= 0.98, catCols)
    
    #We drop the following columns
    colsToBeRemoved = list(filteredCols)
    
    #Delete colsToBeRemoved from the df
    df = drop_col(df, colsToBeRemoved)
    
    # replace ? with nan
    df = df.replace('?',np.nan)
    
    #Delete colsToBeRemoved from the df
    df = drop_col(df, ["weight"])
    
    #Replace NaN values with Unknown in select columns
    col_list = ["tel_2", "tel_1", "race"]
    for col in col_list:
        df[col] = df[col].fillna("Unknown")
    
    #Delete colsToBeRemoved from the df
    df = drop_col(df, ["tel_11","tel_10","tel_9",'patient_id'])
    
    def groupVals(col_list, df):
        for col in col_list:
            valFreq = (df[col].value_counts())/df.shape[0]
    #         print("THis is valFreq: ", valFreq)
            lessFreqVals = valFreq[valFreq<=.01]
    #         print("\n\n")
    #         print("These are values with lessFreq: ", lessFreqVals)        
            df.loc[df[col].isin(lessFreqVals.index.tolist()),col] = "Others"
            return df
    
    df = groupVals(["tel_1","tel_2"], df)
    
#     df = df[~(df["gender"] == "Unknown/Invalid")]

    age_id = {'[0-10)':0, 
          '[10-20)':10, 
          '[20-30)':20, 
          '[30-40)':30, 
          '[40-50)':40, 
          '[50-60)':50,
          '[60-70)':60, 
          '[70-80)':70, 
          '[80-90)':80, 
          '[90-100)':90}
    df['age'] = df.age.replace(age_id)
    
#     df = drop_col(df, ["encounter_id", "patient_id"])
    
    return df
    
    
    
    
```

```python
def dataProcessing(df, scaler):
    
    # Get list of numerical columns
    numCols = df.select_dtypes(include=np.number).columns.tolist()
    
    # Select categorical columns
    catCols = df.select_dtypes("object").columns
    
    #Get dummy variables
#     df = add_dummies(df, catCols, dropx=True)
    for eachCol in catCols:
        df[eachCol] = pd.factorize(df[eachCol])[0] 
    
    if "diabetesMed" in numCols:
        numCols.remove("diabetesMed")
    else:
        pass
    
    #Transform test data
    df[numCols] = scaler.transform(df[numCols])
    
    return df
```
