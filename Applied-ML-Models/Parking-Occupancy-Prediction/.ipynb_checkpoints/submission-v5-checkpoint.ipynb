# 3. Data Cleaning

### Key date cleaning steps perfomed
- Change `LastUpdated` from `string` to a `datetime` type.
- Remove `whitespace` from `SystemCodeNumber` and `OccupancyTrend` columns
- Drop `ExitRate` and `FillRate` columns
- Drop `OccupancyTrend` column
- Remove rows wherein value of `Occupancy > Capacity`. It could be a typing error or sensor measurement error. 
- Update calculation of `OccupancyPercentage` column as the values are incorrect
- Remove rows with `NIA North` and `BHMBRTARC01` SystemCodeNumber values
- Remove `7:30 am timestamps` as we only have 30 values for it
- Create `new date/time` related `columns`
- Check and remove `duplicate values` in the dataset

```python
def data_cleaning(df):
    
    #Convert LastUpdated from string to datetime object
    df["LastUpdated"] = pd.to_datetime(df["LastUpdated"])
    
    # Select categorical columns
    catCols = df.select_dtypes("object").columns
    
    # Remove whitespace from categorical columns
    for col in catCols:

        # applying strip function on column
        df[col] = df[col].map(str.strip)
        
    #Drop ExitRate and FillRate
    df = drop_col(df,["ExitRate","FillRate"])
    
    #Drop OccupancyTrend
    df = drop_col(df,["OccupancyTrend"])
    
    #Look for potential incorrect values in Occupancy Column
    df = df[~(df["Occupancy"] > df["Capacity"])]
    
    # Remove rows wherein Occupancy < 0
    df = df[df["Occupancy"] < 0]
    
    
    
    
```
