# 3. Data Cleaning

### Key date cleaning steps perfomed
- Change `LastUpdated` from `string` to a `datetime` type.
- Remove `whitespace` from `SystemCodeNumber` and `OccupancyTrend` columns
- Drop `ExitRate` and `FillRate` columns
- Drop `OccupancyTrend` column
- Remove rows wherein value of `Occupancy > Capacity`. It could be a typing error or sensor measurement error. 
- Update calculation of `OccupancyPercentage` column as the values are incorrect
- Remove `7:30 am timestamps` as we only have 30 values for it
- Create `new date/time` related `columns`
- Check and remove `duplicate values` in the dataset

```python
df.head()
```

**3.1 Update LastUpdated column Data Type**

```python
#Convert LastUpdated from string to datetime object
df["LastUpdated"] = pd.to_datetime(df["LastUpdated"])
```

**3.1 Remove whitespace from object columns**

```python
df.info()
```

```python
# Select categorical columns
catCols = df.select_dtypes("object").columns
```

```python
# Remove whitespace from categorical columns
for col in catCols:

    # applying strip function on column
    df[col] = df[col].map(str.strip)
```

**3.2 Drop ExitRate and FillRate columns**

We Drop `ExitRate` and `FillRate` columns as we don't have sufficient information on the columns as per the data dictionary.<br>

- As per the data dictionary, we understand
    - ExitRate - Represents the exit rate of cars
    - FillRate - Represents the fill rate of cars

- We can't use this data as we don't know the unit of rate of change, whether it is rate per 10 minutes/15 minutes etc.
<br>
Since, we don't have sufficient information on the columns ExitRate and FillRate, we drop both the columns.

```python
#Drop ExitRate and FillRate
df = drop_col(df,["ExitRate","FillRate"])
```

**3.3 Drop OccupancyTrend Column**

We Drop `OccupancyTrend` column as we don't have sufficient information on the columns as per the data dictionary.

```python
#Drop OccupancyTrend
df = drop_col(df,["OccupancyTrend"])
```

**3.4 Check the values in Occcupancy should always be less than or equal to Capacity**

```python
# Check the range of OccupancyPercentage - should be between 0% and 100%
print('Minimum Percent Occupied: {:.2%}'.format(df["OccupancyPercentage"].min()))
print('Maximum Percent Occupied: {:.2%}'.format(df["OccupancyPercentage"].max()))
```

As the actual OccupancyPercentage goes upto 12,740%, it seems like there is a potential error in the Occupancy field and also in OccupancyPercentage calculations.

- Drop rows from the dataframe wherein Occupancy > Capacity
- Update OccupancyPercentage column with correct calculations

```python
#Look for potential incorrect values in Occupancy Column
df = df[~(df["Occupancy"] > df["Capacity"])]
```

```python
#Check if we still have Occupancy values > Capacity values
df[df["Occupancy"] > df["Capacity"]]
```

**3.5 Update OccupancyPercentage field as the calculations look incorrect**

```python
df['OccupancyPercentage'] = df["Occupancy"] / df["Capacity"]
```

```python
# Re-check range
print('Minimum Percent Occupied: {:.2%}'.format(df["OccupancyPercentage"].min()))
print('Maximum Percent Occupied: {:.2%}'.format(df["OccupancyPercentage"].max()))
```

```python
# Rows wherein OccupancyPercentage < 0
df[df["OccupancyPercentage"] < 0]
```

```python
# Remove rows wherein OccupancyPercentage < 0
df = df[df["OccupancyPercentage"] >= 0]
```

```python
df.head()
```

**3.6 Check whether we have the correct timestamps in the dataset**

```python
df.groupby(df["LastUpdated"].dt.hour)["LastUpdated"].count()
```

**3.7 Create new columns for date time details**

```python
# Create new columns extracting information from the LastUpdated column - 
# month, week, dayofweek, date, date_time_halfhour and time
df['month'] = df.LastUpdated.dt.month
df['week'] = df.LastUpdated.dt.week
df['dayofweek'] = df.LastUpdated.dt.dayofweek
df['date'] = df.LastUpdated.dt.date
df['date_time_halfhour'] = df.LastUpdated.dt.round('30min')
df['time'] = df.date_time_halfhour.dt.time
```

Let's look at the time stamps in the dataset for which we have parking details

```python
df.groupby(["time"]).size()
```

We see that we have only 30 entries for 07:30 am which we will discard.

```python
df[df.time == datetime.time(7,30)]
```

```python
# filter out few 7:30 measurements
df = df[df.time > datetime.time(7,30)]
```

## 5. Data Pre-Processing

### Key Data Pre-Processing steps performed

- Convert LastUpdated to Numeric
- Create Dummy variable for SystemCodeNumber | Month | Week | Day of week
- Extract hours and minutes from time column
- Remove the following columns - "Capacity", "Occupancy", "date" and "date_time_halfhour"
- Split the data into train and test data
- Scale train and test data
- Split X and y from train and test data respectively

**5.1 Set LastUpdated as index column**

```python
df.set_index('LastUpdated', inplace=True)
```

**5.2 Create Dummy variable for SystemCodeNumber | Month | Week | Day of week**

```python
#Get dummy variables
df = add_dummies(df, ["SystemCodeNumber","month","week","dayofweek"], dropx=True)
```

**5.3 Extract hours and minutes from time column**

```python
#Extract hours and minutes from time column
df["time"] = df.apply(lambda x: x["time"].hour + (x["time"].minute)/60, axis=1)
```

**5.4 Remove unwanted columns**

```python
#Drop selected columns
df = drop_col(df, ["date", "date_time_halfhour", "OccupancyPercentage"])
```

**5.5 Split the data into train and test data**

```python
#Split the train and test data
df_train, df_test = df_partitioning(df, test_size=0.3, random_state=23, shuffle=True)
print(df_train.shape)
print(df_test.shape)
```

```python
df_train
```

**5.5 Scale train and test data separately**

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ["time", "Capacity", "Occupancy"]
# df_train, df_test = scaling(df_train, df_test, num_vars)
```

**5.6 Split x and y from scaled train-test set**

```python
y_train = df_train.pop('Occupancy')
X_train = df_train
```
