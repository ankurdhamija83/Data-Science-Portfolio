# 2. Overview of the dataset

**Dataset Overview**
- SystemCodeNumber: ID code for the parking area
- Capacity: The number of parking spots available
- ExitRate: Represents the exit rate of cars
- FillRate: Represents the fill rate of cars
- OccupancyPercentage: Represents the occupancy percentage
- OccupancyTrend: Represents the trend in the car occupancy
- Occupancy: The number of parkings spots actually occupied at the time

# 3. Data Cleaning

### Key steps to be performed on the dataset

- We already know that one thing we need to do is change `LastUpdated` to be a `datetime` type.

After doing so, there are a few additional features to add that will help with our analysis:

1. OccupancyPercentage - Ratio of Occupancy to Capacity. From an initial look at the dataset, the calculation in this column seems to be incorrect. Need to correct the values and also have to make sure these values go from 0%-100%
2. Date - Only the date component of LastUpdated. Used for checking the data.
3. DayOfWeek - An integer representing the day of the week. (1 = Monday)
4. Date_Time_HalfHour - this field will round each time to the nearest half hour. This will help when we aggregate accross all parking areas.
5. Time - Only the time component of the Date_Time_HalfHour. Used for checking the data.

```python
df.head()
```

```python
#Convert LastUpdated from string to datetime object
df["LastUpdated"] = pd.to_datetime(df["LastUpdated"])
```

```python
df.dtypes
```

```python
#Visualize the OccupancyPercentage series
sns.kdeplot(data=df, x="OccupancyPercentage")
```

```python
#Check for values wherein Occupancy is greater than the Capacity
df.loc[df["Occupancy"] > df["Capacity"],:]
```

```python
#Get date, dayofweek, date_time_halfhour and time from LastUpdated column
df['date'] = df["LastUpdated"].dt.date
df['dayofweek'] = df.LastUpdated.dt.dayofweek
df['date_time_halfhour'] = df.LastUpdated.dt.round('30min')
df['time'] = df.date_time_halfhour.dt.time
```
