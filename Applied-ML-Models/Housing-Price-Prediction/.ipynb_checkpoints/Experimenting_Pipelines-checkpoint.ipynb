#### Each Custom Transformer is written as a different class with the following methods
- init()
- fit()
- transform()

## 4. Define Custom Transformers and Pipeline steps

```python
#Custom Transformer that extracts columns passed as argument to its constructor 
class FeatureSelector( BaseEstimator, TransformerMixin ):
    #Class Constructor 
    def __init__( self, feature_names ):
        self._feature_names = feature_names 
    
    #Return self nothing else to do here    
    def fit( self, X, y = None ):
        return self 
    
    #Method that describes what we need this transformer to do
    def transform( self, X, y = None ):
        print("Feature selector transformer called")
        return X[ self._feature_names ] 
```

```python
#Custom Transformer that transforms Categorical columns
class CategoricalTransformer( BaseEstimator, TransformerMixin):
    
    #Class Constructor
    def __init__(self, binaryColList, binaryValList):
        self._binaryColList = binaryColList
        self._binaryValList = binaryValList

    #helper function to do the mapping
    def add_mapping(df, col_list, val_list):

        # Defining the map function
        def binary_map(x):
            return x.map({val_list[0]: 0, val_list[1]: 1})

        # Applying the function to the housing list
        df[col_list] = df[col_list].apply(binary_map)

        return df
        
    #Return self nothing else to do here
    def fit( self, X, y = None  ):
        return self
    
    #Transformer method we wrote for this transformer 
    def transform(self, X , y = None ):
        
        #Do Binary Mapping
        X = add_mapping(X, self._binaryColList, self._binaryValList)
        
        #One-Hot Encoding of categorical columns
        #Get dummy variables
        for eachCol in ["furnishingstatus"]:
            X[eachCol] = pd.factorize(X[eachCol])[0]
        
        print("This is X inside categoricaltransformer: ")
        print(X.head())
        return X
```

```python
#Custom Transformer that transforms Numerical columns
class NumericalTransformer( BaseEstimator, TransformerMixin):
    
    
    #Class Constructor
    def __init__(self):
        print("This is inside NumericalTransformer class")
        pass
    
    #Return self nothing else to do here
    def fit( self, X, y = None  ):
        return self
    
    #Transformer method we wrote for this transformer 
    def transform(self, X , y = None ):
        print("This is X inside numericaltransformer: ", X)
        return X
```

### Create the Pipeline

```python
#Defining the steps in the categorical pipeline 
categorical_pipeline = Pipeline( steps = [ ( 'cat_selector', FeatureSelector(catCols) ),
                                  
                                  ( 'cat_transformer', CategoricalTransformer(binaryColList, binaryValList) ) ] )
```

```python
#Defining the steps in the numerical pipeline     
numerical_pipeline = Pipeline( steps = [ ( 'num_selector', FeatureSelector(numCols) ),
                                  
                                  ( 'num_transformer', NumericalTransformer() ),
                                                                    
                                  ( 'min_max_scaler', MinMaxScaler() ) ] )
```

```python
#Combining numerical and categorical piepline into one full big pipeline horizontally 
#using FeatureUnion
full_pipeline = FeatureUnion( transformer_list = [ ( 'categorical_pipeline', categorical_pipeline ), 
                                                  
                                                  ( 'numerical_pipeline', numerical_pipeline ) ] )
```

```python
#The full pipeline as a step in another pipeline with an estimator as the final step
full_pipeline_m = Pipeline( steps = [ ( 'full_pipeline', full_pipeline),
                                  
                                  ( 'model', linear_model.LinearRegression() ) ] )

#Can call fit on it just like any other pipeline
full_pipeline_m.fit( X_train, y_train )

#Can predict with it like any other pipeline
y_test_pred = full_pipeline_m.predict( X_test ) 
```
